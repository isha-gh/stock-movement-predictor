```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/raw/AAPL_prices.csv')
print(df.info())
print(df.describe())
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df['Date'], df['Close'])
plt.title('APPL Close Price Over Time')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
